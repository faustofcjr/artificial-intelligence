# Regression with Keras

### Predict Global Sales

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold

import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
```

### LOADING DATASET

```python
df = pd.read_csv('data/games_out.csv')
df
```

## Creating Arrays for the Features and the Response Variable

```python
X = df.iloc[:, 5:].values
y = df['Global_Sales'].values
```

## Creating the train and test datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print('X_train.shape=', X_train.shape, '/ X_test.shape=', X_test.shape)
```

## Building the Deep Learning Regression Model

```python
def baseline_model():
    input_shape = len(X[0])
    units = round(input_shape / 2)
    
    model = Sequential()
    model.add(Dense(units=units, input_shape=(input_shape,), activation='relu'))
    model.add(Dense(units=units, activation='relu'))
    model.add(Dense(units=1, activation='relu'))
    
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

## Evaluate model

#### Number of inputs

```python
input_shape = len(X[0])
units = round(input_shape / 2)
print('input_shape=', input_shape, "/ units=", units)
```

```python
estimator = KerasRegressor(build_fn=baseline_model, epochs=10, batch_size=5, verbose=0)
kfold = KFold(n_splits=10)
results = cross_val_score(estimator, X, y, cv=kfold, scoring='neg_mean_squared_error')
```
